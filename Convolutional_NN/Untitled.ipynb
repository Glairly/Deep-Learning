```python
from tqdm import tqdm
import matplotlib as plt
def compress(x,size = 50):
    arrSize = x.size()[0]
    comSize = int(arrSize/(size**2))
    ind = 0
    compressed = []
    while(ind < arrSize):
        z = x[ind:ind+comSize].sum()/comSize
        compressed.append(z)
        ind+=comSize
    compressed = torch.FloatTensor(compressed)
    
    return compressed

    
        
        
```

```python
z = compress(x) 
plt.plot(z)
plt.show()
```
